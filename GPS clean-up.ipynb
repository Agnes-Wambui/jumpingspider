```python
import pandas as pd
import reverse_geocoder
import numpy as np
```

```python
gps = pd.read_csv("C:/Users/sarah/Downloads/gps.csv", sep = "\t")
gps.head()
```

```python
gps[["gps_lat","gps_long","gps_extra"]] = gps.gps.str.split(",", expand = True)
gps["gps_lat"] = gps["gps_lat"].astype(float) 
gps["gps_long"] = gps["gps_long"].astype(float)
gps["gps2"] = list(zip(gps.gps_lat,gps.gps_long))
gps["len"] = len(gps.gps2)
gps.head()
```

```python
gps_list = gps.gps2.values.tolist()
location = reverse_geocoder.search(gps_list)
gps["town"] = [p["name"] for p in location]
gps["county"] = [p["admin1"] for p in location]
gps["country_code"] = [p["cc"] for p in location]
gps
```

```python
gps.to_csv("gps_with_country.csv")
```
